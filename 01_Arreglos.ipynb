## 02 - Recorrido de una lista

La forma de recorrer una lista y una tupla es igual, así que en esta sección nos vamos a centrar en las listas

### Bucle for

En python ***no existe el for tradicional*** (`for(i=0; i < n; i++){...}`) sino que se implementa el ***for each*** que es una variación que facilita el recorrido de elementos en estructuras de datos como listas, tuplas, diccionarios y más. El bucle **for each** itera automáticamente sobre cada elemento en la estructura, eliminando la necesidad de gestionar manualmente el índice o el contador, como se hace en el **for** tradicional.

La sintaxis básica del bucle **for each** en Python es la siguiente:

```python
for elemento in estructura_de_datos:
    # Hacer algo con el elemento
```

Aquí, **elemento** es una variable que tomará el valor de cada elemento en la **estructura_de_datos** en cada iteración del bucle. Por ejemplo, si deseas recorrer una lista de números y mostrar cada número, puedes hacerlo de la siguiente manera:

```python
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(numero)
```

Este código imprimirá los números del 1 al 5, uno en cada línea. El bucle **for each** también puede utilizarse con otras *estructuras de datos*, como diccionarios, donde el **elemento** representará las claves en cada iteración, permitiéndote acceder a los valores correspondientes.

En resumen, el bucle **for each** en Python ofrece una manera más simple y legible de recorrer elementos en estructuras de datos, sin preocuparse por el control manual de los índices. Esto hace que el código sea más limpio y menos propenso a errores.

#### Ejemplos

**Ejemplo 1: Recorrer una lista e imprimir elementos**

````python
frutas = ["manzana", "banana", "kiwi", "naranja"]
for fruta in frutas:
    print(fruta)
````

**Ejemplo 2: Calcular la suma de una lista de números**

````python
numeros = [5, 10, 15, 20, 25]
suma = 0
for numero in numeros:
    suma += numero
print("La suma es:", suma)
````

**Ejemplo 3: Generar una lista de cuadrados**

````python
numeros = [1, 2, 3, 4, 5]
cuadrados = []
for numero in numeros:
    cuadrados.append(numero ** 2)
print("Cuadrados:", cuadrados)
````

## 03 - Algoritmos

### Ordenamiento burbuja

Bubble-sort-example-300px.gif

El algoritmo de ordenación de burbuja es un sencillo método de ordenación que funciona comparando y intercambiando repetidamente elementos adyacentes si están en el orden incorrecto. A medida que se recorre la lista, los elementos más grandes "burbujean" hacia el final de la lista, de ahí su nombre. Aunque es un algoritmo simple, no es muy eficiente en términos de tiempo de ejecución, especialmente para listas grandes, ya que tiene una complejidad de tiempo de **O(n^2)**.

Aquí tienes una implementación en pseudocódigo del algoritmo de ordenación de burbuja:

```
Algoritmo Burbuja:
    Entrada: Una lista de elementos a ordenar
    Salida: La lista ordenada

    Para i desde 0 hasta longitud(lista) - 1:
        Para j desde 0 hasta longitud(lista) - i - 1:
            Si lista[j] > lista[j+1] Entonces
                Intercambiar lista[j] y lista[j+1]

Fin Algoritmo
```

En este pseudocódigo, `lista` representa la lista de elementos que deseamos ordenar. La idea principal es recorrer la lista varias veces, comparando cada par de elementos adyacentes y realizando un intercambio si es necesario para asegurar que estén en el orden correcto.

En la implementación, el primer bucle `para` controla cuántas pasadas se hacen a través de la lista. Cada vez que se completa una pasada, el elemento más grande ya está en su posición correcta al final de la lista, por lo que no es necesario compararlo nuevamente en las pasadas siguientes. Por eso, el segundo bucle `para` recorre solo hasta `longitud(lista) - i - 1`.

Si el elemento en la posición `j` es mayor que el elemento en la posición `j+1`, se realiza un intercambio para que queden en el orden correcto. El proceso se repite en cada pasada hasta que la lista esté completamente ordenada.

Aunque el algoritmo de burbuja es fácil de entender e implementar, su rendimiento es limitado para tareas de ordenación en listas grandes debido a su complejidad cuadrática. Hay algoritmos más eficientes, como el **QuickSort** o el MergeSort, que se utilizan en aplicaciones reales para ordenar grandes conjuntos de datos de manera más rápida.
